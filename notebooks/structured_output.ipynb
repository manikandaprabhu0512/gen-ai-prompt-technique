```python
from pydantic import BaseModel, Field
from typing import List
```

```python
class ResponseStructure(BaseModel) :
    name: str = Field(description="Name of the person")
    email: str = Field(description="Email of the person")
    age: int = Field(description="Age of the person")

structured_llm = llm.with_structured_output(ResponseStructure)

res = structured_llm.invoke(text)

print(res.name)
```

```python
class Movies(BaseModel) :
    name : str = Field(descripition="Name of the movie")
    year : int = Field(description="Year of the movie")
    rating : float = Field(description="Rating of the movie")

class AllMovies(BaseModel):
    movies : List[Movies]

structured_llm = llm.with_structured_output(AllMovies)
```
