```python
from dotenv import load_dotenv
load_dotenv()
```

```python
documents = [
    "The food was delicious and the waiter...",
    "The service was terrible and the food was bad...",
    "The food was good and the waiter was nice...",
    "I did not like the food and the waiter was bad...",
]
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
vector = embeddings.embed_documents(documents)
```

```python
len(vector[0])
```

### ChromaDB (Vector Embedding -> Vector Store)

```python
from langchain_chroma import Chroma
```

```python
vector_store = Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="./vector_db"
)
```

```python
query = "Nice service & great food"
result = vector_store.similarity_search(query, k=2)
print(result)
```
