```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
        it will return the sum of 2 numbers.

        Args:
            a: number one
            b: number two
    """

    return a+b

@tool
def multiply_numbers(a:int, b:int):
    """
        it will return the product of 2 numbers.

        Args:
            a: number one
            b: number two
    """

    return a*b
```

```python
add_numbers.invoke({"a" : 5, "b" : 6})
```

```python
from langchain.agents import create_agent
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-20b")
```

```python
agent = create_agent(
    model = llm, 
    tools=[add_numbers, multiply_numbers],
    system_prompt="You're a math teacher, and always use tools for calculation."
)
```

```python
response = agent.invoke({"messages":[{"role":"user", "content":"What is the product of 30 and 6?"}]})
```
