### Prompt Templates - Dynammic Templates

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
out = StrOutputParser()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {
        "role": "system",
        "content": "You are a helpful assistant that translates English to {language}."
    },
    {
        "role": "user",
        "content": "{query}"
    }
])
```

```python
final_prompt = prompts.format_messages(language="Tamlish", query="How are you?")
```

```python
res = llm.invoke(final_prompt)
```

```python
print(res.content)
```

```python
res = llm.invoke(prompts.invoke({"language":"Kannada", "query" : "I love youuuuu."}))
```

```python
print(res.content)
```

```python
def transform(result : str):
    return result.upper()
```

```python
chains = prompts | llm | out | transform
```

```python
res = chains.invoke({"language":"Hinglish", "query" : "I love youuuuu."})
```

```python
print(res)
```
