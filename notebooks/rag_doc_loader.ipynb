### TEXT Loader

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/speech.txt")
texts = loader.load()

oneDoc = texts[0]
print(oneDoc.page_content)
```

### PDF Doc Loader

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("../data/medical_report.pdf")
docs = loader.load()
```

```python
oneDoc = docs[1]
print(oneDoc.page_content)
```

### CSV File Loader

```python
from langchain_community.document_loaders import CSVLoader
loader = CSVLoader("../data/employee_records.csv")
data = loader.load()
```

```python
oneData = data[0]
print(oneData.page_content)
```

### WEB Loader

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(web_path="https://en.wikipedia.org/wiki/Deep_Learning")
```

```python
docs = loader.load()
print(docs[0].metadata)
```

### Wiki Loader

```python
from langchain_community.document_loaders import WikipediaLoader
loader = WikipediaLoader(query="Deep Learning", load_max_docs=2)
docs = loader.load()
print(docs[0].page_content)
```
